```python
class Bottle(object): #Python 2~ object was necessary
    def __init__(self, quantity):
        self.__quantity = quantity

    def fill(self, quantity):
        if self.__quantity + quantity > 100:
            self.__quantity = 100
        else:
            self.__quantity += quantity
    def show(self):
      print(self.__quantity)
# __~==> private parameters, must start with __init__ and self for parameters and functions
class CokeBottle(Bottle):
  def __init__(self,quantity,taste):
    super().__init__(quantity) #use parent's class init
    self.__taste=taste
  def show(self):
    super().show() #use parent's class init
    print(self.__taste)

btl1=Bottle(30)
btl2=CokeBottle(20,'chilling') #inherited Bottle => same method can be used including __init__/=
btl1.fill(30)
btl2.fill(70)
btl1.show()
btl2.show()
```

```python
class Employee(object):
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __str__(self):
        return 'name : {}, salary = {}'.format(self.name, self.salary)

    def __repr__(self):
        return 'name : {}, salary = {}'.format(self.name, self.salary)

    def __add__(self, other):
        if type(other) == int:
            new_emp = Employee(self.name, self.salary + other)
        else:
            new_emp = Employee(self.name + other.name, self.salary + other.salary)
        return new_emp

    def __radd__(self, other):
        return Employee(self.name, self.salary + other)

    def __len__(self):
        return len(self.name)

    def __gt__(self, other): # greater than
        print ("__gt__ is called")
        return self.salary > other.salary

    def __gte__(self, other): # equal
        print ("__gte__ is called")
        return self.salary >= other.salary

    def __lt__(self, other): # less than, it appears 5 times as
        print ("__lt__ is called")
        return self.salary < other.salary

    def __lte__(self, other):
        print ("__lte__ is called")
        return self.salary <= other.salary

    def __eq__(self, other):
        print ("__eq__ is called")
        return len(self.name) == len(other.name)
e1 = Employee('bob', 1000)
e2 = Employee('kate', 2000)
e3 = Employee('ken', 2500)
e4 = Employee('fox', 500)
print(e1 + e2)
print(e1 + e2 + e3)
print(e1 + 1000) #this will call '__add__', As intergers will be added to salary for __add__
print(1000 + e1)
print(len(e1))
print ("="*50)
lst = [e1, e4, e2, e3]
print(e1 < e2)
print ()
print(sorted(lst)) #to sort list, calculate e1>e2, e2>e3....=>5 times 'less than' < was called
print ()
print(sorted(lst, key=lambda x : x.salary)) #the key was given by x.salary so no other metods are needed to sort
print ()
print(e1 == e3)
print(e1 == e2)
```

Using a hyphen (-) between the two characters in [ ] means the range (From - To) between the two characters. For example, the regular expression [a-c] is the same as [abc], and [0-5] is the same as [012345].

^

When ^ metacharacter is used inside [], it has the meaning of "not". The regular expression [^0-9] matches only non-numeric character

***"Capitalization of below becomes "not"~"***

**\d**

same as the [0-9]

**\s**

matches whitespace character same as the [ \t\n\r\f\v]

**\w**

same as the [a-zA-Z0-9_]  (number+numeric)


**Asterisk(*)**

means that the character immediately preceding the * can berepeated from 0 to infinity

**+**

means that the character immediately preceding the + can be repeated from 1 to infinity

### Quantile-Quantile Plot
- 수집 데이터를 표준정규분포이 분위수와 비교하여 그리는 그래프 --> 데이터의 정규성 가정에 대한 검토 가능하게 함

Another popular plot for checking the distribution of a data sample is the quantile-quantile plot, Q-Q plot, or QQ plot for short.

This plot generates its own sample of the idealized distribution that we are comparing with, in this case the Gaussian distribution. The idealized samples are divided into groups (e.g. 5), called quantiles. Each data point in the sample is paired with a similar member from the idealized distribution at the same cumulative distribution.

The resulting points are plotted as a scatter plot with the idealized value on the x-axis and the data sample on the y-axis.

A perfect match for the distribution will be shown by a line of dots on a 45-degree angle from the bottom left of the plot to the top right. Often a line is drawn on the plot to help make this expectation clear. Deviations by the dots from the line shows a deviation from the expected distribution.


```python
# QQ Plot
from numpy.random import seed
from numpy.random import randn
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot
# seed the random number generator
seed(122)
# generate univariate observations
data = 5 * randn(100) + 50
# q-q plot
qqplot(data, line='s')
pyplot.show()
```

```python
# Shapiro-Wilk Test
from numpy.random import seed
from numpy.random import randn
from scipy.stats import shapiro
# seed the random number generator
seed(1)
# generate univariate observations
data = 5 * randn(100) + 50
# normality test
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')

# D'Agostino and Pearson's Test
from numpy.random import seed
from numpy.random import randn
from scipy.stats import normaltest
# seed the random number generator
seed(1)
# generate univariate observations
data = 5 * randn(100) + 50
# normality test
stat, p = normaltest(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

#### Shapiro-Wilk vs. D'Agostino K^2 Test
- Shapiro-Wilk
  - 샘플 크기: Shapiro-Wilk 검정은 작은 샘플 크기에 적합하며 일반적으로 50개 미만의 샘플에 가장 효과적(그러나 최대 2000개의 샘플까지도 사용할 수 있음)
  - 민감도: 이 검정은 데이터의 정규성에 대해 매우 민감. 즉, 작은 표본에서도 비정규성을 잘 감지할 수 있음
  - 사용 상황: 작은 데이터셋을 사용할 때 또는 정규성이 중요한 정밀한 분석을 수행할 때 적합
- D'Agostino K^2 Test
  - 통계적 척도: 이 검정은 왜도와 첨도라는 두 가지 통계적 척도를 사용하여 데이터의 정규성을 평가(왜도는 대칭성의 부족을, 첨도는 꼬리의 두께를 측정)
  - 샘플 크기: D'Agostino의 K^2 검정은 상대적으로 큰 샘플 크기에 더 적합할 수 있음
  - 사용 상황: 데이터의 왜도나 첨도가 중요한 역할을 할 때, 또는 좀 더 큰 샘플 크기를 다룰 때 유용
