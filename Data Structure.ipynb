```python
import sys
from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def arraytoBST(self, arr):
        if not arr:
            return None
        mid = len(arr) // 2
        root = Node(arr[mid])
        root.left = self.arraytoBST(arr[:mid])
        root.right = self.arraytoBST(arr[mid+1:])
        return root

    def add(self, root, data):
        new_node = Node(data)
        if self.root is None:
            self.root = new_node
            return True
        if root.data < data:
            if root.right is not None:
                return self.add(root.right, data)
            else:
                root.right = new_node
                return True
        elif root.data > data:
            if root.left is not None:
                return self.add(root.left, data)
            else:
                root.left = new_node
                return True
        return False
    def delete(self, root, key):
      if not root:
          return root
      elif key < root.data:
          root.left = self.delete(root.left, key)
      elif key > root.data:
          root.right = self.delete(root.right, key)
      else:
        # Node with only one child or no child
        if not root.left:
            return root.right
        elif not root.right:
            return root.left

        # Node with two children: Get the inorder successor (smallest
        # in the right subtree)
        temp = self.get_min_value_node(root.right)

        # Copy the inorder successor's content to this node
        root.data = temp.data

        # Delete the inorder successor
        root.right = self.delete(root.right, temp.data)

      return root
    def get_min_value_node(self, node):
      current = node

    # Loop down to find the leftmost leaf
      while current.left is not None:
        current = current.left
      return current

    def search(self, root, data):
        if self.root is None:
            print("Tree is empty")
            return False
        if root.data == data:
            print("Found!")
            return True
        elif root.data > data:
            return self.search(root.left, data)
        else:
            return self.search(root.right, data)

    def inorder(self, root):
        if self.root is None:
            print("empty")
            return False
        if root is None:
            return
        self.inorder(root.left)
        print(root.data)
        self.inorder(root.right)

    def preorder(self, root):
        if self.root is None:
            print("empty")
            return False
        if root is None:
            return
        print(root.data)
        self.preorder(root.left)
        self.preorder(root.right)

    def postorder(self, root):
        if self.root is None:
            print("empty")
            return False
        if root:
            self.postorder(root.left)
            self.postorder(root.right)
            print(root.data)

    def _getHeight(self, curr):
        if not curr:
            return 0
        return 1 + max(self._getHeight(curr.left), self._getHeight(curr.right))

    def _printSpaces(self, n, curr):
        width = len(str(curr.data))  # Get the width of the data
        for _ in range(int(n * width)):
            print(" ", end="")
        if not curr:
            print(" ", end="")
        else:
            print(str(curr.data), end="")

    def printTree(self):
        if not self.root:
            return

        q = deque()
        q.append(self.root)
        temp = deque()
        cnt = 0
        height = self._getHeight(self.root) - 1
        nMaxNodes = 2 ** (height + 1) - 1

        while cnt <= height:
            curr = q.popleft()
            if len(temp) == 0:
                self._printSpaces(nMaxNodes / (2 ** (cnt + 1)), curr)
            else:
                self._printSpaces(nMaxNodes / (2 ** cnt), curr)
            if not curr:
                temp.append(None)
                temp.append(None)
            else:
                temp.append(curr.left)
                temp.append(curr.right)
            if not q:
                print("\n")
                q = temp
                temp = deque()
                cnt += 1

# Example usage:
tree = BST()
this = [i for i in range(1, 10)]
print(this)
tree.root = tree.arraytoBST(this)
print("postorder")
tree.postorder(tree.root)
print("inorder")
tree.inorder(tree.root)
print("preorder")
tree.preorder(tree.root)
tree.delete(tree.root,6)
print("preorder")
tree.preorder(tree.root)
```

```python
MAX_CAPACITY = 1000
class MaxHeap:
    def __init__(self, num=MAX_CAPACITY):
        self.elements = [0] * num
        self.count = 0
        self.capacity = num
    def parent(self, i):
        if i <= 0 or i >= self.count:
            return -1
        return (i - 1) // 2
    def left(self, i):
        left_child_index = 2 * i + 1
        if left_child_index >= self.count:  # No left child
            return -1
        return left_child_index
    def right(self, i):
        right_child_index = 2 * i + 2
        if right_child_index >= self.count:  # No right child
            return -1
        return right_child_index
    def insertElement(self, elem):
        if self.count >= self.capacity:  # Check if the heap is full
            raise Exception("Exception: Heap is full.")
        self.count += 1
        i = self.count - 1
        self.elements[i] = elem

        while i != 0 and self.elements[self.parent(i)] < self.elements[i]: #if the value is bigger than parent's one
            self.elements[i], self.elements[self.parent(i)] = self.elements[self.parent(i)], self.elements[i] #switch
            i = self.parent(i) #in order to re-test, i should indicate "new element's location"
    def maximum(self):
        if self.count == 0: #Exception Handling, Can't find maximum if count==0 as it means que is empty.
            raise Exception("Exception: Heap is empty.")
        return self.elements[0]
    def deleteMaximum(self):
        if self.count == 0: #Exception Handling, Can't delete maximum if count==0 as it means que is empty.
            raise Exception("Exception: Heap is empty.")
        if self.count == 1: #If que holds only one value, that value is maximum itself. And no need to handle more as count==0.
            self.count -= 1 #So, just remove it from heap.
            return self.elements[0]
        #As count>1, we should set new "king" that will replace eliminated "maximum"
        root = self.elements[0]
        self.elements[0] = self.elements[self.count - 1]
        self.count -= 1
        self.maxHeapify(0) #After removing maximum, make sure heap keeps its properties.
    def maxHeapify(self, i):
        largest = i
        l = self.left(i) #Finding left child's index
        r = self.right(i) #Finding right child's index
        if l != -1 and l < self.count and self.elements[l] > self.elements[largest]: #Check if left child is bigger than the parent
            largest = l
        if r != -1 and r < self.count and self.elements[r] > self.elements[largest]: #Check if right child is bigger than the parent
            largest = r
        if largest != i: #If one of the children is bigger, than parent, swap them.
            self.elements[i], self.elements[largest] = self.elements[largest], self.elements[i]
            self.maxHeapify(largest)

         #Return the modified heap in order to assign new root.

    def buildHeap(self, array):
        self.count = len(array)
        self.capacity = max(self.capacity, self.count)
        self.elements = array
        for i in range(self.count // 2 - 1, -1, -1):
            self.maxHeapify(i)

```

```python
import sys
import heapq
from collections import deque
import math

CONSTRUCT = 'C'
IS_ADJACENT = 'I'
GET_NEIGHBORS = 'N'
TOPOLOGICAL_SORT = 'T'
SHORTEST_PATH = 'S'

class Graph:
    def __init__(self):
        self.adj_list = {}

    def construct_graph(self, n, m, edges):
        self.adj_list = {i: [] for i in range(n)} #setting vertex_ids
        for u, v in edges: #u: start point v: end point,weight
            self.adj_list[u].append(v)

    def is_adjacent(self, u, v):
        if v not in self.adj_list or u not in self.adj_list: #if u, v is not in vertex_ids => Raise error
            raise Exception("Invalid vertex_ids")
        u_list=[i[0] for i in self.adj_list[u]]  #list of all points
        if v in u_list:
            return 'T'
        else:
            return 'F'
    def get_neighbors(self, u):
        if u not in self.adj_list: #if u is not in vertex_ids => Raise error
            raise Exception("Invalid vertex_id")
        return [i[0] for i in self.adj_list[u]] #return 'list' of neighbors

    def topological_sort(self):
        indegree = {v: 0 for v in self.adj_list}
        #This is for Indegree of each vertex
        for neighbors in self.adj_list.values():
            for neighbor, _ in neighbors:
                indegree[neighbor] += 1
        queue = deque([v for v in self.adj_list if indegree[v] == 0])
        #First, deque contains vertices with 0 indegree
        top_order = []
        while queue: #while queue is not empty
            u = queue.popleft() #draw leftmost one.
            top_order.append(u) #append to result
            for v, _ in self.adj_list[u]: #_ is distance, so we don't use for this method.


                indegree[v] -= 1 #As we check degree by only "Indegree", this code is to consider "outdegree" (Indegree involve outdegree)
                if indegree[v] == 0: #This means, v has lowest degree around remains.
                    queue.append(v)

        if len(top_order) != len(self.adj_list): #when top_order's length is not same with total number of vertices, we should know that topological_sort didn't perform well and there was a cycle of the graph
            raise Exception("Graph contains a cycle")
        return top_order
    def bfs(self, start):
        visited = set()
        queue = [start]
        bfs_order = []

        while queue:
            vertex = queue.pop(0)
            if vertex not in visited:
                visited.add(vertex)
                bfs_order.append(vertex)
                queue.extend([neighbor for neighbor in self.adj_list[vertex] if neighbor not in visited])

        return bfs_order

    def dfs(self, start):
        visited = set()
        stack = [start]
        dfs_order = []

        while stack:
            vertex = stack.pop()
            if vertex not in visited:
                visited.add(vertex)
                dfs_order.append(vertex)
                stack.extend([neighbor for neighbor in self.adj_list[vertex] if neighbor not in visited])

        return dfs_order


    def shortest_path(self,start,end):
        dist = {v: math.inf for v in self.adj_list} #Dijkstra Algorithm: All infinity distance at first
        dist[start] = 0 #Except for start point itself
        prev = {v: None for v in self.adj_list}
        visited = set()
        Q = [(0, start)] #each tuple consists of 1. distance accumlated 2.point traversed recently
        while Q:
            current_dist,u = heapq.heappop(Q)
            if u in visited: #u was visited, so skip it.
                continue
            visited.add(u) #Mark u as visited
            if u == end: #when we reached the end point.
                path = []
                while u is not None: #updating path by referring "previous" vertex.
                    path.append(u)
                    u = prev[u]

                return list(reversed(path)) #we should reverse path as it stored endpoint for first and updated itself by referring "previous" vertex.

            for v, weight in self.adj_list[u]: #check neighbors of u
                if current_dist + weight < dist[v]: #If new path is shorter
                    dist[v] = current_dist + weight #Update accumlated distance
                    prev[v] = u #Update previous=current
                    heapq.heappush(Q,(dist[v],v)) #push updated one


        return None
```

```python
def selection_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Find the minimum element in the remaining unsorted array
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Swap the found minimum element with the first element
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

# Example usage
arr = [64, 25, 12, 22, 11]
selection_sort(arr)
print("Sorted array is:", arr)
def insertion_sort(arr):
    n = len(arr)
    # Traverse through 1 to len(arr)
    for i in range(1, n):
        key = arr[i]
        # Move elements of arr[0..i-1], that are greater than key, to one position ahead of their current position
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        # Place the key in its correct position
        arr[j + 1] = key

# Example usage
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print("Sorted array is:", arr)
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Divide the array into two halves
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Recursively sort each half
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # Merge the sorted halves
    sorted_arr = merge(left_half, right_half)

    return sorted_arr

def merge(left, right):
    sorted_arr = []
    left_index = right_index = 0

    # Merge two sorted arrays into sorted_arr
    while left_index < len(left) and right_index < len(right):
        if left[left_index] <= right[right_index]:
            sorted_arr.append(left[left_index])
            left_index += 1
        else:
            sorted_arr.append(right[right_index])
            right_index += 1

    # Append remaining elements
    sorted_arr.extend(left[left_index:])
    sorted_arr.extend(right[right_index:])

    return sorted_arr

# Example usage:
arr = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
sorted_arr = merge_sort(arr)
print("Sorted array:", sorted_arr)
```

```python
def partition(arr, low, high):
    pivot=arr[high]  # Pick a pivot from arr
    left,right=low,high
    while left < right:
        while left<right and arr[left]>pivot:  # Change for descending order
            left += 1
        while left<right and arr[right]<=pivot:  # Change for descending order
            right -= 1
        if left <right:
            arr[left],arr[right]=arr[right],arr[left]
    arr[high],arr[left]=arr[left],pivot
    return left

def quicksort(arr,low,high):
    if low<high:
        p=partition(arr,low,high)
        quicksort(arr,low, p-1)
        quicksort(arr,p+1, high)
    return arr
arr = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
sorted_arr = quicksort(arr, 0, len(arr) - 1)
print("Sorted array:", sorted_arr)
```

```python
class Node:
    def __init__(self):
        self.children={}
        self.is_end=False

class Trie:
    def __init__(self):
        self.root=Node()

    def insert(self,word):
        node=self.root
        for ch in word:
            if ch not in node.children:
                node.children[ch]=Node()
            node=node.children[ch]
        node.is_end=True

    def search_prefix(self,prefix):
        node=self.root
        for ch in prefix:
            if ch not in node.children:
                return []
            node = node.children[ch]

        return self.find(node,prefix)

    def find(self,node,prefix):
        words = []
        if node.is_end:
            words.append(prefix)

        for ch, word in node.children.items():
            words.extend(self.find(word, prefix+ch))

        return words
```
